```python
%%writefile Named_Entity_Recognition.py

import streamlit as st 
import spacy
from spacy import displacy 

st.title ('Named Enitity Recognition Application') 
text = st.text_area ('Enter your text here') 

nlp_model = spacy.load('en_core_web_sm')

def Text_Ext(text):
    document = nlp_model(text)
    entities = [(ent.text , ent.label_) for ent in document.ents]
    return entities

def show_entities(document):
  json = displacy.render(document, style = 'ent', jupyter = False)
  return json

if st.button ('Extract Here'):
    if text:
        document = nlp_model(text)
        entites = Text_Ext(text)
        st.write(entites)
        st.subheader('Extracted Entites')
        st.markdown(show_entities(document), unsafe_allow_html= True)
```
